```python
import pandas as pd
import numpy as np

# ============================================
# COMPLETE STUDENT ANALYSIS SYSTEM
# ============================================

np.random.seed(42)

data = {
    'Name': ['Ali','Sara','Usman','Fatima','Bilal',
             'Zara','Ahmed','Hina','Kamran','Nadia'],
    'City': ['Karachi','Lahore','Islamabad','Karachi','Lahore',
             'Islamabad','Karachi','Lahore','Karachi','Islamabad'],
    'Math':    [85, 90, 78, 92, 88, 76, 95, 82, 70, 88],
    'Science': [80, 85, 72, 88, 91, 70, 90, 78, 65, 92],
    'English': [78, 92, 80, 85, 76, 88, 82, 90, 72, 85],
}

df = pd.DataFrame(data)

# --- Average marks calculate karo ---
df['Average'] = df[['Math','Science','English']].mean(axis=1)
df['Average'] = df['Average'].round(2)

# --- Grade assign karo ---
def assign_grade(avg):
    if avg >= 90:   return 'A+'
    elif avg >= 80: return 'A'
    elif avg >= 70: return 'B'
    else:           return 'C'

df['Grade'] = df['Average'].apply(assign_grade)

# --- Pass/Fail ---
df['Status'] = df['Average'].apply(
    lambda x: 'Pass' if x >= 60 else 'Fail'
)

print("=" * 60)
print("         COMPLETE STUDENT REPORT")
print("=" * 60)
print(df.to_string(index=False))

print("\n" + "=" * 60)
print("CITY WISE PERFORMANCE")
print("=" * 60)
city_report = df.groupby('City')['Average'].agg(
    ['mean','max','min','count']
).round(2)
city_report.columns = ['Avg','Highest','Lowest','Students']
print(city_report)

print("\n" + "=" * 60)
print("TOP 3 STUDENTS")
print("=" * 60)
top3 = df.nlargest(3, 'Average')[['Name','City','Average','Grade']]
print(top3.to_string(index=False))

print("\n" + "=" * 60)
print("GRADE DISTRIBUTION")
print("=" * 60)
print(df['Grade'].value_counts())
```
