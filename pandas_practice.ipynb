```python
# TOPIC 1 — Series (Single Column)

import pandas as pd

# ============================================
# SERIES — One column data structure
# ============================================

# Making Simple series 
marks = pd.Series([85, 90, 78, 92, 88])
print(marks)
# 0    85
# 1    90
# 2    78
# 3    92
# 4    88
# dtype: int64

# Left side = INDEX (automatically 0,1,2...)
# Right side = VALUES

# ============================================
# Custom Indexing
# ============================================
marks_named = pd.Series(
    [85, 90, 78, 92, 88],
    index=['Ali', 'Sara', 'Usman', 'Fatima', 'Bilal']
)
print(marks_named)
# Ali      85
# Sara     90
# Usman    78
# Fatima   92
# Bilal    88
# dtype: int64

# Name by index access karo
print(marks_named['Ali'])     # 85
print(marks_named['Fatima'])  # 92

# ============================================
# Series of Properties
# ============================================
print(marks_named.values)   # [85 90 78 92 88]
print(marks_named.index)    # Index(['Ali','Sara'...])
print(marks_named.dtype)    # int64
print(marks_named.shape)    # (5,)

# ============================================
# Series pe Math
# ============================================
print(marks_named.mean())   # 86.6
print(marks_named.max())    # 92
print(marks_named.min())    # 78
print(marks_named.sum())    # 433

# Filter karo
print(marks_named[marks_named >= 88])
# Sara      90
# Fatima    92
# Bilal     88
```

```python
# TOPIC 4 — Columns & Rows Selection

import pandas as pd

data = {
    'Name':   ['Ali','Sara','Usman','Fatima','Bilal'],
    'Age':    [20, 22, 21, 23, 20],
    'City':   ['Karachi','Lahore','Islamabad','Karachi','Lahore'],
    'Marks':  [85, 90, 78, 92, 88],
    'Grade':  ['B', 'A', 'C', 'A', 'B']
}
df = pd.DataFrame(data)

# ============================================
# COLUMN SELECT KARNA
# ============================================

# Single column — Series milti hai
names = df['Name']
print(names)

# Multiple columns — DataFrame milta hai
subset = df[['Name', 'Marks', 'Grade']]
print(subset)

# ============================================
# ROW SELECT KARNA — 2 tarike
# ============================================

# .iloc[] — Index NUMBER se (i = integer)
print(df.iloc[0])      # Pehli row
print(df.iloc[-1])     # Aakhri row
print(df.iloc[1:4])    # Row 1 se 3 tak
print(df.iloc[0:3, 0:2])  # Rows 0-2, Columns 0-1

# .loc[] — NAAM se (label based)
print(df.loc[0])           # Row of Index 0 
print(df.loc[0:2])         # Index of 0 to 2
print(df.loc[0:2, 'Name':'Marks'])  # Rows + column names

# ============================================
# iloc vs loc — Farq Yaad Karo!
# ============================================
# iloc → Integer location → using numbers
# loc  → Label location  → using names

# Trick of remembering:
# iloc → "i" = integer = number
# loc  → label = name
```

```python
#  TOPIC 5 — Filtering (Mostly used in DataFrames)

import pandas as pd

data = {
    'Name':   ['Ali','Sara','Usman','Fatima','Bilal','Zara','Ahmed'],
    'Age':    [20, 22, 21, 23, 20, 24, 21],
    'City':   ['Karachi','Lahore','Islamabad','Karachi',
                'Lahore','Islamabad','Karachi'],
    'Marks':  [85, 90, 78, 92, 88, 76, 95],
    'Grade':  ['B', 'A', 'C', 'A', 'B', 'C', 'A']
}
df = pd.DataFrame(data)

# ============================================
# FILTERING — Condition se rows nikalo
# ============================================

# 85 se zyada marks wale
high_marks = df[df['Marks'] > 85]
print("High Marks:\n", high_marks)

# Students of karachi
karachi = df[df['City'] == 'Karachi']
print("\nKarachi Students:\n", karachi)

# ============================================
# Multiple Conditions
# ============================================

# AND condition — & use here
# Karachi + marks > 80
karachi_good = df[(df['City'] == 'Karachi') & (df['Marks'] > 80)]
print("\nKarachi + Good Marks:\n", karachi_good)

# OR condition — | use here
# Lahore ya Islamabad
not_karachi = df[(df['City'] == 'Lahore') | (df['City'] == 'Islamabad')]
print("\nLahore or Islamabad:\n", not_karachi)

# isin() — Multiple values ek saath check
cities = df[df['City'].isin(['Karachi', 'Lahore'])]
print("\nKarachi or Lahore (isin):\n", cities)

# Grade A wale
grade_a = df[df['Grade'] == 'A']
print("\nGrade A Students:\n", grade_a)
```

```python
# Topic 6 Adding or Modifying new columns

import pandas as pd

data = {
    'Name':   ['Ali','Sara','Usman','Fatima','Bilal'],
    'Marks':  [85, 90, 78, 92, 88],
}
df = pd.DataFrame(data)

# ============================================
# NEW COLUMN ADD KARNA
# ============================================

# Simple new column
df['Passed'] = True
print(df)

# Calculation se column
df['Marks_Out_50'] = df['Marks'] / 2
print(df)

# Condition se column
df['Grade'] = df['Marks'].apply(
    lambda x: 'A' if x >= 90
    else 'B' if x >= 80
    else 'C'
)
print(df)

# ============================================
# COLUMN DELETE 
# ============================================
df.drop('Marks_Out_50', axis=1, inplace=True)
# axis=1 = column delete karna
# inplace=True = original df mein change karo

print(df)

# ============================================
# RENAMING COLUMN
# ============================================
df.rename(columns={'Marks': 'Total_Marks'}, inplace=True)
print(df.columns)
# ['Name', 'Total_Marks', 'Passed', 'Grade']
```

```python
#  TOPIC 7 — Missing Data Handling (Real World!)

import pandas as pd
import numpy as np

# DATA IS ALWAYS INCOMPLETE IN REAL WORLD!
data = {
    'Name':   ['Ali', 'Sara', 'Usman', 'Fatima', 'Bilal'],
    'Age':    [20, np.nan, 21, np.nan, 20],      # 2 missing
    'Marks':  [85, 90, np.nan, 92, 88],           # 1 missing
    'City':   ['Karachi', 'Lahore', None, 'Karachi', 'Lahore']
}
df = pd.DataFrame(data)
print("Original Data:")
print(df)
#      Name   Age  Marks     City
# 0     Ali  20.0   85.0  Karachi
# 1    Sara   NaN   90.0   Lahore
# 2   Usman  21.0    NaN     None
# 3  Fatima   NaN   92.0  Karachi
# 4   Bilal  20.0   88.0   Lahore

# ============================================
# MISSING VALUES DHUNDHO
# ============================================
print("\nKahan missing hai:")
print(df.isnull())
#     Name    Age  Marks   City
# 0  False  False  False  False
# 1  False   True  False  False
# 2  False  False   True   True
# 3  False   True  False  False
# 4  False  False  False  False

print("\nKitne missing hain har column mein:")
print(df.isnull().sum())
# Name     0
# Age      2
# Marks    1
# City     1

# ============================================
# MISSING VALUES HANDLING
# ============================================

# Option 1 — Drop karo (missing wali rows hatao)
df_dropped = df.dropna()
print("\nAfter dropna:")
print(df_dropped)   # only complete rows

# Option 2 — Fill it (With mean value)
df['Age'].fillna(df['Age'].mean(), inplace=True)
print("\nAge filled with mean:")
print(df['Age'])
# 0    20.0
# 1    20.33  ← mean se fill hua
# 2    21.0
# 3    20.33  ← mean se fill hua
# 4    20.0

# Option 3 — Fill it (With fixed value)
df['City'].fillna('Unknown', inplace=True)
print("\nCity filled:")
print(df['City'])
```

```python
# TOPIC 8 — GroupBy (Powerful Analysis!)

import pandas as pd

data = {
    'Name':   ['Ali','Sara','Usman','Fatima','Bilal','Zara','Ahmed'],
    'City':   ['Karachi','Lahore','Islamabad','Karachi',
                'Lahore','Islamabad','Karachi'],
    'Marks':  [85, 90, 78, 92, 88, 76, 95],
    'Grade':  ['B', 'A', 'C', 'A', 'B', 'C', 'A']
}
df = pd.DataFrame(data)

# ============================================
# GROUPBY — Category wise analysis
# ============================================

# Har city ka average marks
city_avg = df.groupby('City')['Marks'].mean()
print("City wise Average Marks:")
print(city_avg)
# City
# Islamabad    77.0
# Karachi      90.67
# Lahore       89.0

# Har city mein kitne students
city_count = df.groupby('City')['Name'].count()
print("\nCity wise Student Count:")
print(city_count)
# Karachi      3
# Lahore       2
# Islamabad    2

# Har grade ka max marks
grade_max = df.groupby('Grade')['Marks'].max()
print("\nGrade wise Max Marks:")
print(grade_max)

# Multiple aggregations ek saath
city_stats = df.groupby('City')['Marks'].agg(
    ['mean', 'max', 'min', 'count']
)
print("\nCity wise Complete Stats:")
print(city_stats)
```

```python
#  TOPIC 9 — Sorting

import pandas as pd

data = {
    'Name':   ['Ali','Sara','Usman','Fatima','Bilal'],
    'Age':    [20, 22, 21, 23, 20],
    'Marks':  [85, 90, 78, 92, 88],
}
df = pd.DataFrame(data)

# Marks se sort karo — ascending (low to high)
df_sorted = df.sort_values('Marks')
print("Marks ascending:\n", df_sorted)

# Marks se sort karo — descending (high to low)
df_sorted_desc = df.sort_values('Marks', ascending=False)
print("\nMarks descending:\n", df_sorted_desc)

# Multiple columns se sort
df_multi = df.sort_values(['Age', 'Marks'], ascending=[True, False])
print("\nAge then Marks:\n", df_multi)

# Index reset karo after sorting
df_sorted_desc = df_sorted_desc.reset_index(drop=True)
print("\nReset index:\n", df_sorted_desc)
```
