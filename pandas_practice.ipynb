```python
# TOPIC 1 — Series (Single Column)

import pandas as pd

# ============================================
# SERIES — One column data structure
# ============================================

# Making Simple series 
marks = pd.Series([85, 90, 78, 92, 88])
print(marks)
# 0    85
# 1    90
# 2    78
# 3    92
# 4    88
# dtype: int64

# Left side = INDEX (automatically 0,1,2...)
# Right side = VALUES

# ============================================
# Custom Indexing
# ============================================
marks_named = pd.Series(
    [85, 90, 78, 92, 88],
    index=['Ali', 'Sara', 'Usman', 'Fatima', 'Bilal']
)
print(marks_named)
# Ali      85
# Sara     90
# Usman    78
# Fatima   92
# Bilal    88
# dtype: int64

# Name by index access karo
print(marks_named['Ali'])     # 85
print(marks_named['Fatima'])  # 92

# ============================================
# Series of Properties
# ============================================
print(marks_named.values)   # [85 90 78 92 88]
print(marks_named.index)    # Index(['Ali','Sara'...])
print(marks_named.dtype)    # int64
print(marks_named.shape)    # (5,)

# ============================================
# Series pe Math
# ============================================
print(marks_named.mean())   # 86.6
print(marks_named.max())    # 92
print(marks_named.min())    # 78
print(marks_named.sum())    # 433

# Filter karo
print(marks_named[marks_named >= 88])
# Sara      90
# Fatima    92
# Bilal     88
```

```python
# TOPIC 2 — DataFrame (Poori Table)

import pandas as pd

# ============================================
# Making DataFrame  — From Dictionary 
# ============================================

data = {
    'Name':    ['Ali', 'Sara', 'Usman', 'Fatima', 'Bilal'],
    'Age':     [20, 22, 21, 23, 20],
    'City':    ['Karachi', 'Lahore', 'Islamabad', 'Karachi', 'Lahore'],
    'Marks':   [85, 90, 78, 92, 88],
    'Passed':  [True, True, True, True, True]
}

df = pd.DataFrame(data) # Converting dictionary to DataFrame
print(df)
```

```python
# TOPIC 3 — DataFrame Ko Explore Karo

import pandas as pd

data = {
    'Name':   ['Ali','Sara','Usman','Fatima','Bilal','Zara','Ahmed'],
    'Age':    [20, 22, 21, 23, 20, 24, 21],
    'City':   ['Karachi','Lahore','Islamabad','Karachi',
                'Lahore','Islamabad','Karachi'],
    'Marks':  [85, 90, 78, 92, 88, 76, 95],
    'Grade':  ['B', 'A', 'C', 'A', 'B', 'C', 'A']
}

df = pd.DataFrame(data)

# ============================================
# EXPLORATION — Pehli cheez jo hamesha karo!
# ============================================

# Pehli 5 rows dekho
print("HEAD:")
print(df.head())        # default = 5 rows
print(df.head(3))       # sirf 3 rows

# Aakhri 5 rows dekho
print("\nTAIL:")
print(df.tail())        # default = 5 rows
print(df.tail(2))       # sirf 2 rows

# Shape — kitni rows, kitne columns
print("\nShape:", df.shape)    # (7, 5)

# Column names
print("Columns:", df.columns.tolist())
# ['Name', 'Age', 'City', 'Marks', 'Grade']

# Data types of each column 
print("\nData Types:")
print(df.dtypes)
# Name     object   (string)
# Age       int64   (number)
# City     object   (string)
# Marks     int64   (number)
# Grade    object   (string)

# Summary statistics — everything in one line!
print("\nStatistics:")
print(df.describe())
#        Age      Marks
# count  7.0      7.000
# mean  21.57    86.28
# std    1.39     6.87
# min   20.0     76.00
# 25%   20.5     82.50
# 50%   21.0     88.00
# 75%   22.5     91.00
# max   24.0     95.00

# Complete info — missing values bhi dikhata hai
print("\nInfo:")
print(df.info())
```

```python
# TOPIC 4 — Columns & Rows Selection

import pandas as pd

data = {
    'Name':   ['Ali','Sara','Usman','Fatima','Bilal'],
    'Age':    [20, 22, 21, 23, 20],
    'City':   ['Karachi','Lahore','Islamabad','Karachi','Lahore'],
    'Marks':  [85, 90, 78, 92, 88],
    'Grade':  ['B', 'A', 'C', 'A', 'B']
}
df = pd.DataFrame(data)

# ============================================
# COLUMN SELECT KARNA
# ============================================

# Single column — Series milti hai
names = df['Name']
print(names)

# Multiple columns — DataFrame milta hai
subset = df[['Name', 'Marks', 'Grade']]
print(subset)

# ============================================
# ROW SELECT KARNA — 2 tarike
# ============================================

# .iloc[] — Index NUMBER se (i = integer)
print(df.iloc[0])      # Pehli row
print(df.iloc[-1])     # Aakhri row
print(df.iloc[1:4])    # Row 1 se 3 tak
print(df.iloc[0:3, 0:2])  # Rows 0-2, Columns 0-1

# .loc[] — NAAM se (label based)
print(df.loc[0])           # Row of Index 0 
print(df.loc[0:2])         # Index of 0 to 2
print(df.loc[0:2, 'Name':'Marks'])  # Rows + column names

# ============================================
# iloc vs loc — Farq Yaad Karo!
# ============================================
# iloc → Integer location → using numbers
# loc  → Label location  → using names

# Trick of remembering:
# iloc → "i" = integer = number
# loc  → label = name
```

```python
#  TOPIC 5 — Filtering (Mostly used in DataFrames)

import pandas as pd

data = {
    'Name':   ['Ali','Sara','Usman','Fatima','Bilal','Zara','Ahmed'],
    'Age':    [20, 22, 21, 23, 20, 24, 21],
    'City':   ['Karachi','Lahore','Islamabad','Karachi',
                'Lahore','Islamabad','Karachi'],
    'Marks':  [85, 90, 78, 92, 88, 76, 95],
    'Grade':  ['B', 'A', 'C', 'A', 'B', 'C', 'A']
}
df = pd.DataFrame(data)

# ============================================
# FILTERING — Condition se rows nikalo
# ============================================

# 85 se zyada marks wale
high_marks = df[df['Marks'] > 85]
print("High Marks:\n", high_marks)

# Students of karachi
karachi = df[df['City'] == 'Karachi']
print("\nKarachi Students:\n", karachi)

# ============================================
# Multiple Conditions
# ============================================

# AND condition — & use here
# Karachi + marks > 80
karachi_good = df[(df['City'] == 'Karachi') & (df['Marks'] > 80)]
print("\nKarachi + Good Marks:\n", karachi_good)

# OR condition — | use here
# Lahore ya Islamabad
not_karachi = df[(df['City'] == 'Lahore') | (df['City'] == 'Islamabad')]
print("\nLahore or Islamabad:\n", not_karachi)

# isin() — Multiple values ek saath check
cities = df[df['City'].isin(['Karachi', 'Lahore'])]
print("\nKarachi or Lahore (isin):\n", cities)

# Grade A wale
grade_a = df[df['Grade'] == 'A']
print("\nGrade A Students:\n", grade_a)
```
