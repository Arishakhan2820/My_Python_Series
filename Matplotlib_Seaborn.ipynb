```python
# Professional Plot — With all options

import matplotlib.pyplot as plt
import numpy as np

# ============================================
# PROFESSIONAL LINE PLOT
# ============================================

months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']
sales  = [150, 180, 120, 200, 250, 230,
          280, 300, 270, 310, 290, 350]

# Figure size set (width, height in inches)
plt.figure(figsize=(12, 5))

# Plot karo — styling ke saath
plt.plot(
    months,          # x axis
    sales,           # y axis
    color='blue',    # line color
    linewidth=2,     # line boldness
    marker='o',      # circle at every point
    markersize=8,    # size of the circle
    linestyle='-',   # solid line
    label='Sales'    # name for legend
)

# Labels aur Title
plt.title('Monthly Sales 2024', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales (Units)', fontsize=12)

# Grid — lines which help in reading the graph
plt.grid(True, alpha=0.3)  # alpha = transparency

# Legend — line ka naam dikhao
plt.legend(fontsize=12)

# Layout theek karo
plt.tight_layout()

# Show karo
plt.show()
```

```python
# TOPIC 2 — Bar Chart (ForComparison)

import matplotlib.pyplot as plt
import numpy as np

# ============================================
# VERTICAL BAR CHART
# ============================================

products = ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Charger']
sales    = [745, 930, 830, 1230, 560]
colors   = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6']

plt.figure(figsize=(10, 6))

bars = plt.bar(products, sales, color=colors,
               width=0.6, edgecolor='black', linewidth=0.5)

# Har bar ke upar value dikhao
for bar, value in zip(bars, sales):
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # x position
        bar.get_height() + 10,               # y position
        str(value),                          # text
        ha='center',                         # horizontal align
        fontsize=11,
        fontweight='bold'
    )

plt.title('Annual Product Sales', fontsize=14, fontweight='bold')
plt.xlabel('Products')
plt.ylabel('Total Units Sold')
plt.grid(axis='y', alpha=0.3)   # sirf horizontal grid
plt.tight_layout()
plt.show()

# ============================================
# HORIZONTAL BAR CHART — barh()
# ============================================
plt.figure(figsize=(10, 6))

plt.barh(products, sales, color=colors,
         edgecolor='black', linewidth=0.5)

plt.title('Annual Product Sales (Horizontal)',
          fontsize=14, fontweight='bold')
plt.xlabel('Total Units Sold')
plt.ylabel('Products')
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# TOPIC 3 — Histogram ( Watching Distribution)

import matplotlib.pyplot as plt
import numpy as np

# ============================================
# HISTOGRAM — Data ka distribution dekhna
# AI mein hamesha use karte hain!
# ============================================

np.random.seed(42)
# 200 students ke marks
marks = np.random.normal(loc=75, scale=10, size=200)
# normal distribution — average 75, std 10

plt.figure(figsize=(10, 5))

plt.hist(
    marks,
    bins=20,           # kitne bars banao
    color='steelblue',
    edgecolor='black',
    alpha=0.7,
    label='Students'
)

# Average line dikhao
avg = np.mean(marks)
plt.axvline(avg, color='red', linewidth=2,
            linestyle='--', label=f'Average: {avg:.1f}')

plt.title('Student Marks Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# bins = zyada → more detail
# bins = kam   → overall shape
```

```python
#  TOPIC 4 — Scatter Plot (Relationship Dekhna)

import matplotlib.pyplot as plt
import numpy as np

# ============================================
# SCATTER PLOT
# Do variables ka relationship dekhna
# AI mein features ka relation samjhna
# ============================================

np.random.seed(42)
study_hours = np.random.uniform(1, 10, 50)
marks       = study_hours * 8 + np.random.normal(0, 5, 50)
marks       = np.clip(marks, 0, 100)  # 0-100 ke beech rakho

plt.figure(figsize=(9, 6))

scatter = plt.scatter(
    study_hours,
    marks,
    c=marks,           # color acc to marks
    cmap='RdYlGn',     # Red-Yellow-Green colormap
    s=100,             # point size
    alpha=0.8,
    edgecolors='black',
    linewidth=0.5
)

# Colorbar
plt.colorbar(scatter, label='Marks')

# Trend line add karo
z = np.polyfit(study_hours, marks, 1)
p = np.poly1d(z)
x_line = np.linspace(1, 10, 100)
plt.plot(x_line, p(x_line), color='blue',
         linewidth=2, linestyle='--', label='Trend Line')

plt.title('Study Hours vs Marks', fontsize=14, fontweight='bold')
plt.xlabel('Study Hours per Day')
plt.ylabel('Marks (%)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()


```

```python
# TOPIC 5 — Pie Chart

import matplotlib.pyplot as plt

# ============================================
# PIE CHART — Percentage distribution
# ============================================

grades  = ['A+', 'A', 'B', 'C', 'Fail']
counts  = [15, 25, 30, 20, 10]
colors  = ['#2ecc71','#3498db','#f39c12','#e67e22','#e74c3c']
explode = [0.1, 0, 0, 0, 0.1]  # A+ aur Fail thoda bahar

plt.figure(figsize=(8, 8))

wedges, texts, autotexts = plt.pie(
    counts,
    labels=grades,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',    # percentage dikhao
    startangle=90,        # 12 baje se shuru
    shadow=True           # shadow effect
)

# Text size
for text in texts:
    text.set_fontsize(12)
for autotext in autotexts:
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')

plt.title('Grade Distribution', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# TOPIC 6 — Subplots (Multiple Graphs together)

import matplotlib.pyplot as plt
import numpy as np

# ============================================
# SUBPLOTS — Multiple graphs ek figure mein
# Professional reports mein hamesha use hota!
# ============================================

np.random.seed(42)
months = ['Jan','Feb','Mar','Apr','May','Jun']
sales  = [150, 180, 120, 200, 250, 230]
marks  = np.random.normal(75, 10, 100)
products = ['Laptop','Phone','Tablet','Headphones']
units    = [745, 930, 830, 1230]

# 2x2 grid of plots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Business Dashboard', fontsize=16, fontweight='bold')

# Plot 1 — Line Chart (top left)
axes[0, 0].plot(months, sales, color='blue',
                marker='o', linewidth=2)
axes[0, 0].set_title('Monthly Sales')
axes[0, 0].set_xlabel('Month')
axes[0, 0].set_ylabel('Units')
axes[0, 0].grid(True, alpha=0.3)

# Plot 2 — Histogram (top right)
axes[0, 1].hist(marks, bins=15, color='green',
                edgecolor='black', alpha=0.7)
axes[0, 1].set_title('Marks Distribution')
axes[0, 1].set_xlabel('Marks')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].grid(True, alpha=0.3)

# Plot 3 — Bar Chart (bottom left)
axes[1, 0].bar(products, units,
               color=['#3498db','#e74c3c','#2ecc71','#f39c12'])
axes[1, 0].set_title('Product Sales')
axes[1, 0].set_xlabel('Product')
axes[1, 0].set_ylabel('Units Sold')
axes[1, 0].grid(axis='y', alpha=0.3)

# Plot 4 — Pie Chart (bottom right)
axes[1, 1].pie(units, labels=products,
               autopct='%1.1f%%', startangle=90)
axes[1, 1].set_title('Market Share')

plt.tight_layout()
plt.show()
```

```python
#  TOPIC 7 — Seaborn (Pretty Graphs, less Code!)

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# ============================================
# SEABORN SETUP
# ============================================

# Style set karo — automatically sundar dikhta hai!
sns.set_theme(style='whitegrid', palette='muted')

# Student data
np.random.seed(42)
data = {
    'Name':    [f'Student_{i}' for i in range(50)],
    'City':    np.random.choice(['Karachi','Lahore','Islamabad'], 50),
    'Math':    np.random.randint(60, 100, 50),
    'Science': np.random.randint(55, 100, 50),
    'English': np.random.randint(60, 100, 50),
    'Grade':   np.random.choice(['A','B','C'], 50)
}
df = pd.DataFrame(data)
df['Average'] = df[['Math','Science','English']].mean(axis=1).round(1)
```

```python
# Topic 8 Seaborn Plot Types

# ============================================
# 1. BOXPLOT — Distribution + Outliers
# ============================================
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='City', y='Average',
            palette='Set2', width=0.5)
plt.title('City wise Marks Distribution', fontsize=14)
plt.xlabel('City')
plt.ylabel('Average Marks')
plt.tight_layout()
plt.show()

# Box ka matlab:
# Middle line = Median
# Box edges   = 25% aur 75%
# Whiskers    = Min aur Max
# Dots        = Outliers (bahut alag values)

# ============================================
# 2. VIOLIN PLOT — Distribution ka shape
# ============================================
plt.figure(figsize=(10, 6))
sns.violinplot(data=df, x='City', y='Average',
               palette='Set3')
plt.title('City wise Score Violin Plot', fontsize=14)
plt.tight_layout()
plt.show()

# ============================================
# 3. HEATMAP — Correlation Matrix (AI mein!)
# ============================================
plt.figure(figsize=(8, 6))

# Correlation calculate karo
corr = df[['Math', 'Science', 'English', 'Average']].corr()

sns.heatmap(
    corr,
    annot=True,       # numbers dikhao
    fmt='.2f',        # 2 decimal places
    cmap='coolwarm',  # color scheme
    vmin=-1, vmax=1,  # range -1 to 1
    square=True,
    linewidths=0.5
)
plt.title('Subject Correlation Heatmap', fontsize=14)
plt.tight_layout()
plt.show()

# Correlation:
#  1.0 = perfect positive relation
# -1.0 = perfect negative relation
#  0.0 = koi relation nahi

# ============================================
# 4. PAIRPLOT — Sab variables ek saath!
# ============================================
sns.pairplot(
    df[['Math','Science','English','Average','Grade']],
    hue='Grade',       # Grade ke hisaab se color
    palette='Set1',
    diag_kind='kde',   # diagonal pe distribution
    plot_kws={'alpha': 0.6}
)
plt.suptitle('Complete Data Pairplot', y=1.02, fontsize=14)
plt.tight_layout()
plt.show()
# Yeh AI mein features ka relationship dekhne ke liye use hota hai!

# ============================================
# 5. COUNTPLOT — Category count
# ============================================
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='City', hue='Grade',
              palette='Set2')
plt.title('City wise Grade Distribution', fontsize=14)
plt.xlabel('City')
plt.ylabel('Count')
plt.legend(title='Grade')
plt.tight_layout()
plt.show()
```
